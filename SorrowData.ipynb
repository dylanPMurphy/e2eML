```python
import pandas as pd
import os

SUICIDE_PATH = os.path.join("datasets", "suicide")

def load_suicide_data(suicide_path=SUICIDE_PATH):
    csv_path = os.path.join(suicide_path, "master.csv")
    return pd.read_csv(csv_path)
```

```python
sorrowTable = load_suicide_data()
```
